```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
files = ["events_PureSM.csv","events_PureAQGC_1em11.csv","events_PureAQGC_1em12.csv", "events_Int_1em12.csv","events_Int_1em11.csv"]
```

```python
data = []
cols = []
allcols = pd.read_csv(files[0], delimiter=" ").columns
name = allcols[0]
index = 1
while name != "w_Default":
    cols.append(name)
    name = allcols[index]
    index+=1
    
for file in files:
    data.append(pd.read_csv(file, delimiter=" "))
```

```python
#1em11
for col in cols:
    vSM,bSM,_ = plt.hist(data[0][col], bins=100, label="sm")
    vP11,bP11,_ = plt.hist(data[1][col], bins=100, label="pure aQGC 1em11")
    #vP12,bP12,_ = plt.hist(data[2][col], bins=100, label="pure aQGC 1em12")
    vI12,bI12,_ = plt.hist(data[3][col], bins=100, label="pure interference 1em11")
    #vI11,bI11,_ = plt.hist(data[4][col], bins=100, label="pure interference 1em12")

    plt.xlabel(col)
    plt.ylabel('occurence')
    plt.title('Overlayed Plots')
    plt.legend()
    plt.show()
```
